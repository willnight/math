```python
X = np.linspace(-5 * np.pi, 5 * np.pi, 300)

def my_plot(pl, x, k, a, b, color):
    pl.plot(x, k * np.cos(x - a) + b, color = color)

plt.figure(figsize = (10, 5))
my_plot(plt, X, 2, 1, 0, 'r')
my_plot(plt, X, 18, 5, 4, 'b')
my_plot(plt, X, 8, 1, 7, 'g')

plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

x = np.linspace(-2, 3, 201)

plt.figure(figsize = (12, 12))
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 6) 
plt.grid(True)
plt.axis('scaled')
plt.show()

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2.5, 6.5))
x2, y2 =  fsolve(equations, (-1, 1))
x3, y3 =  fsolve(equations, (5, 15))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
