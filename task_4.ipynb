```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_fun(k):    
    x = np.linspace(-10, 10, 201)
    plt.plot(x, np.cos(k * x), label = f'line {k}')

for i in [1, 3]:
    x = my_fun(i)  
    
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('tight')
plt.show()
```
